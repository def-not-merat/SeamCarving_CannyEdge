```python
import kornia
import torch
import cv2
import matplotlib.pyplot as plt
```

```python
#read image
img_read = plt.imread('bowl-of-fruit.jpg')
ten_bgr = kornia.image_to_tensor(img_read/255)
img = kornia.color.rgb_to_grayscale(ten_bgr)
plt.imshow(kornia.tensor_to_image(img), cmap = 'gray')
img = torch.reshape(img, (1,1,img.shape[1],-1))
```

```python
#return the derivative of a guassioan. 
#since its seprable, return the seperated vectors
def seprable_gausian_deriv(size, sig):
    if size%2 ==0:
        print("size in not odd")
        return
    x = torch.zeros(size)
    for i in range(size):
        x[i] = i-size//2
    x_2 = torch.square(x)
    e_stuff = torch.exp( -x_2 / (2 * sig**2))
    
    deriv_x = -x/(2*torch.pi * sig**4) * e_stuff

    x_comp = torch.reshape(deriv_x, (1,size))
    y_comp = torch.reshape(e_stuff, (1,size))
    return x_comp, y_comp
        
```

```python
#given an angle, return the x, and y associaated to move a pixel in that direction
def dir(theta):
    pi_8 = torch.pi / 8
    if -pi_8<=theta and theta<pi_8:
        return 1,0
    elif pi_8<=theta and theta<3*pi_8:
        return 1,1
    elif 3*pi_8<=theta and theta<5*pi_8:
        return 0,1
    elif 5*pi_8<=theta and theta<7*pi_8:
        return -1,1
    elif -3*pi_8<=theta and theta<-pi_8:
        return 1,-1
    elif -5*pi_8<=theta and theta<-3*pi_8:
        return 0,-1
    elif -7*pi_8<=theta and theta<-5*pi_8:
        return -1,-1
    else:
        return -1,0
```

```python
# canny, without the last step
def MyCanny(grayscale_img, sigma, gradiant_mag_thresh):
    size = int(3*sigma)
    if size%2==0:
        size += 1
    x_comp, y_comp = seprable_gausian_deriv(size, sigma)
    
    #keep in mind, filter2d_seprable is convolution. so we need to filp the kernel
    img_dx = kornia.filters.filter2d_separable(grayscale_img, -x_comp, y_comp)
    img_dy = kornia.filters.filter2d_separable(grayscale_img, -y_comp, x_comp)
    img_grad_dir = torch.atan2(img_dy, img_dx)
    img_grad_mag = torch.sqrt(torch.square(img_dx) + torch.square(img_dy))
    
    img_grad_low = img_grad_mag.detach().clone()
    img_grad_low[img_grad_mag < gradiant_mag_thresh] = 0 

    non_max_suppresion = torch.zeros(img_grad_mag.shape).float()
    
    h = img_grad_mag.shape[2]
    w = img_grad_mag.shape[3]

    for x in range(w):
        for y in range(h):
            if img_grad_low[0,0,y,x] == 0:
                continue
            d_x,d_y = dir(img_grad_dir[0,0,y,x])
            if y+d_y>=h or y+d_y<0 or x+d_x>=w or x+d_x<0 or y-d_y>=h or y-d_y<0 or x-d_x>=w or x-d_x<0:
                non_max_suppresion[0,0,y,x] = img_grad_low[0,0,y,x]
            elif img_grad_low[0,0,y+d_y,x+d_x] <= img_grad_low[0,0,y,x] and img_grad_low[0,0,y-d_y,x-d_x] <= img_grad_low[0,0,y,x]:
                non_max_suppresion[0,0,y,x] = img_grad_low[0,0,y,x]
    return non_max_suppresion, img_dx, img_dy
```

```python
#canny, including last step
def MyCannyFull(grayscale_img, sigma, gradiant_mag_low_thresh, gradiant_mag_high_thresh):
    all_edges,dx,dy = MyCanny(grayscale_img, sigma, gradiant_mag_low_thresh)

    queue = []
    h = all_edges.shape[2]
    w = all_edges.shape[3]
    
    discovered = torch.zeros(all_edges.shape)
    super = all_edges>gradiant_mag_high_thresh
    discovered[super] = 1
    
    for i in range (h):
        for j in range (w):
            if super[0,0,i,j]:
                queue.insert(0,(i,j))

    while queue:
        y,x = queue.pop(0)
        for i in [-1,0,1]:
            for j in [-1,0,1]:
                if i==0 and j==0:
                    continue
                if i+y>=h or i+y<0:
                    continue
                if j+x>=w or j+x<0:
                    continue
                if all_edges[0,0,i+y,j+x]==0:
                    continue
                if discovered[0,0,i+y,j+x]==1:
                    continue
                else:
                    discovered[0,0,i+y,j+x]=1
                    queue.append((i+y,j+x))
    return discovered,all_edges
```
